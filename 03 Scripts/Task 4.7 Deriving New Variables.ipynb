#### Make a subset

```python
#create subset of first one million rows
df = ords_prods_merge[:1000000]
```

```python
#check shape
df.shape
```

```python
#create price_labels
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#look at value counts
df['price_range'].value_counts(dropna = False) 
```

#### Observation 
The top three busiests times of day are 10 am, 11 am, and 2 pm. The least busiest times are 3 am, 4 am, and 2 am. I will need to divide the hours into 3 groups: busiest, average, and fewest.

```python
# creating a for loop
result = []
for value in ords_prods_merge['order_hour_of_day']:
    if value == 10 or value == 11 or value == 14 or value == 13 or value == 15 or value == 12 or value == 16 or value == 9:
        result.append("Most orders")
    elif value == 23 or value == 6 or value == 0 or value == 1 or value == 5 or value == 2 or value == 4 or value  == 3:
        result.append("Fewest orders")
    else:
        result.append("Average orders") 
```

```python
#Create new column 'busiest_period_of_day'
ords_prods_merge['busiest_period_of_day'] = result 
```
