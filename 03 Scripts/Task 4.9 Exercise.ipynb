#### different chart types

```python
#create a bar chart 
day_counts = ords_prods_merge['orders_day_of_week'].value_counts().sort_index()
day_counts.index = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
day_counts.plot.bar(
    xlabel='Day of Week',
    ylabel='Number of Orders',
    title='Orders by Day of Week'
) 
```

```python
# Create a bar chart with day names
day_counts = ords_prods_merge['orders_day_of_week'].value_counts().sort_index()
day_counts.index = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
ax = day_counts.plot.bar(
    xlabel='Day of Week',
    ylabel='Number of Orders',
    title='Orders by Day of Week'
)

# Export figure
ax.figure.savefig(r'C:\Users\hardy\10-2025 Instacart Basket Analysis\04 Analysis\Visualizations\bar_orders_dow.png') 
```
