### Wrangle new dataframe

```python
#change column names to be consistent with other dataframes used. 
df_customers.rename(columns = {'First Name': 'first_name', 'Surnam': 'last_name', 'Gender': 'gender', 'STATE': 'state', 'Age': 'age', 'n_dependants': 'number_of_dependents'}, inplace = True) 
```

```python
#Check output
df_customers
```

```python
#drop first_name and last_name columns, as they do not add anything to my analysis
df_customers = df_customers.drop(columns=['first_name', 'last_name']) 
```

```python
#check output
df_customers.head() 
```

```python
#change column name 'n_dependants' to 'number_of_dependants'
df_customers.rename(columns = {'n_dependents' : 'number_of_dependents'}, inplace = True) 
```

```python
#rename 'fam_status' to 'marital_status'
df_customers.rename(columns = {'fam_status' : 'marital_status'}, inplace = True) 
```

```python
#check outputs
df_customers.head() 
```

All column heads are now in the same format and have names that make sense. Additionally, I removed first and last names as to protect the privacy of the customers as we have a user_id assigned to designate different customers.
